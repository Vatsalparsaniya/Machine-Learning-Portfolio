```python
newgroup = flight.groupby(["year"])
```

```python
total_passengers_of_year_list = []
total_year_list = []
for name,sub_group in newgroup:
            total_passengers_of_year_list.append(sub_group['passengers'].sum())
            total_year_list.append(name)
our_data = pd.DataFrame({'year':total_year_list,'passengers':total_passengers_of_year_list})
our_data.set_index('year')
```

```python
g = sns.pointplot(data = our_data , x="year",y="passengers")
g.set_xticklabels(labels = total_year_list,rotation=45)
```

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
sinplot()
```
